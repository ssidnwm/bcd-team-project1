## 라이브러리 임포트

```python
import os
import json
import pprint as ppr
```

## 필요 함수 정의

```python
def jsonlFinder(directory):
    
    jsonl_files = []
    
    for root, dirs, files in os.walk(directory):
        for file in files:
            if file.endswith('.jsonl'):
                jsonl_files.append(os.path.join(root, file))
            
    return jsonl_files
```

```python
def jsonlMerger(file_list, filePath_output):
    with open(filePath_output, 'w') as outfile:
        for file_name in file_list:
            with open(file_name, 'r') as infile:
                for line in infile:
                    outfile.write(line)
```

## 데이터 파일 경로 지정

```python
data_path_out = r"G:\내 드라이브\LAB_works\법률 LLM 프로젝트\data\데이터 전처리\3. JSON 컨버팅\JSONL_merged_v2"
fileName = "\R078r3e_merged.jsonl"
jsonlMerger(list_jsonlFiles, data_path_out+fileName)
```
